```python
# ------------------ 3. Univariate Analysis ------------------
# Loan Amount Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['loan_amount'], kde=True)
plt.title('Loan Amount Distribution')
plt.show()
```

```python
# ------------------ 3. Univariate Analysis ------------------
# Loan Amount Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['loan_amount'], kde=True)
plt.title('Loan Amount Distribution')
plt.show()
```

```python
# Interest Rate Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['int_rate'], kde=True)
plt.title('Interest Rate Distribution')
plt.show()
```

```python
# Loan Status Count
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='loan_status')
plt.title('Loan Status Distribution')
plt.show()
```

```python
# Home Ownership
plt.figure(figsize=(6,4))
df['home_ownership'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Home Ownership Distribution')
plt.ylabel('')
plt.show()
```

```python
# Purpose
plt.figure(figsize=(10,5))
sns.countplot(data=df, y='purpose', order=df['purpose'].value_counts().index)
plt.title('Loan Purpose Distribution')
plt.show()
```

```python
# ------------------ 4. Bivariate Analysis ------------------
# Loan Amount vs Total Payment
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='loan_amount', y='total_payment', hue='loan_status')
plt.title('Loan Amount vs Total Payment')
plt.show()
```

```python
# Interest Rate vs Loan Status
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='loan_status', y='int_rate')
plt.title('Interest Rate by Loan Status')
plt.show()
```

```python
# DTI vs Loan Status
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='loan_status', y='dti')
plt.title('DTI by Loan Status')
plt.show()
```

```python
# Purpose vs Loan Status
plt.figure(figsize=(10,5))
sns.countplot(data=df, y='purpose', hue='loan_status', order=df['purpose'].value_counts().index)
plt.title('Loan Status by Purpose')
plt.show()
```

```python
# Home Ownership vs Loan Status
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='home_ownership', hue='loan_status')
plt.title('Loan Status by Home Ownership')
plt.show()
```

```python
# ------------------ 5. Time Series Trend ------------------
df['issue_date'] = pd.to_datetime(df['issue_date'], errors='coerce')
df['issue_month'] = df['issue_date'].dt.to_period('M')
```

```python
monthly_trend = df.groupby('issue_month').size()
plt.figure(figsize=(10,5))
monthly_trend.plot()
plt.title('Loan Applications Over Time')
plt.xlabel('Issue Month')
plt.ylabel('Number of Applications')
plt.show()
```

```python
# ------------------ 6. Correlation Heatmap ------------------
plt.figure(figsize=(8,6))
sns.heatmap(df[['loan_amount', 'total_payment', 'int_rate', 'dti']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
